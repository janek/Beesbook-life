### Reading from saved CSV files

```python
#read and interpret timestamp as date
start_csv_name = '2016-08-23_12:00:00.csv'


print('Processing '+start_csv_name)
df = pd.read_csv(start_csv_name, parse_dates=['timestamp'])


#TODO: include dtype conversion
#read and concat a number of hour-long csvs (memory crashes if >16)
for i in range(13,24):
    if i < 10:
        str_i = '0'+str(i)
    else:
        str_i = str(i)
    csv_name = '2016-08-23_'+str_i+':00:00.csv'
    print('Processing '+csv_name)
    new_data = pd.read_csv(csv_name, parse_dates=['timestamp'])
    df = pd.concat([df, new_data])
    print('Num. rows after appending: '+str(df.shape[0]))
```

```python
print(df.shape)
```

```python
num_intervals = 720
interval_length = observ_period // num_intervals

# create zeros in the shape required by (bees x num_invervals)
intervals = pd.DataFrame(data=np.zeros([len(group_ids),num_intervals])) 
ids = pd.DataFrame(data={'id': group_ids})
ids = pd.concat([ids, intervals], axis=1)
ids.head()
```

```python
interval_starttime = datetime_start
ones_count = 0
for i in range(num_intervals): 
    #check detections
    interval_endtime = interval_starttime + interval_length
    before = df['timestamp'] >= interval_starttime 
    after = df['timestamp'] < interval_endtime
    interval_detections = df[before & after]
    n = 0
    for b in ids['id']:
        if b in interval_detections['bee_id'].unique():
            ids.set_value(n, i, 1)
            ones_count += 1
        n += 1 
    interval_starttime = interval_endtime
print(ones_count)
```
